```python
transactions =[]
for i in range(0,len(data)):
  transactions.append([str(data.values[i,j])for j in range(0,20) if str(data.values[i,j]!='0')])
```

```python
!pip install apyori
```

```python
from apyori import apriori
```

```python
rules=apriori(transactions = transactions,min_support=0.003,min_confidence=0.2,min_lift=3)
```

```python
results=list(rules)
results
```

```python
def inspect(results):
  lhs=[tuple(result[2][0][0])[0] for result in results]
  rhs=[tuple(result[2][0][1])[0] for result in results]
  supports=[result[1]for result in results]
  confidences=[result[2][0][2] for result in results]
  lifts=[result[2][0][3]for result in results]
  return list(zip(lhs,rhs,supports,confidences,lifts))
resultsinDataFrame = pd.DataFrame(inspect(results),columns=['Left hand side','Right hand time','Support','Confidence','Lift'])
```
